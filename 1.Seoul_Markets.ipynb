```python
######################## 불러오기 
df_b= pd.read_csv('./data/df_b.csv',encoding='utf-8')   # 발달상권 
df_g= pd.read_csv('./data/df_g.csv',encoding='utf-8')   # 골목상권 

print(df_b.shape)       # 발달상권 
print(df_g.shape)       # 골목상권 
```

# 1. 행정동별 매출금액과 건수 비교 

```python
############################################"발달상권" 행정동 기준, '분기'별 금액 합계 
bal_each_money_sum = df_b.groupby(['행정동', '분기'])['금액'].sum().unstack()

##################### 바 차트 생성
bal_each_money_sum.plot(kind='bar', figsize=(20, 5))
plt.xlabel('자치구')
plt.ylabel('금액 합계')
plt.title('(발달상권) 행정동의 분기별 매출금액 합계')
plt.xticks(rotation=0)  # X축 라벨의 회전 없이 표시
plt.legend(title='분기', labels=['1', '2', '3', '4', '5'])
plt.show()
```

```python
############################################ "골목상권" 행정동 기준, '분기'별 금액 합계 
gol_each_money_sum = df_g.groupby(['행정동', '분기'])['금액'].sum().unstack()

##################### 바 차트 생성
gol_each_money_sum.plot(kind='bar', figsize=(20, 5))
plt.xlabel('자치구')
plt.ylabel('금액 합계')
plt.title('(골목상권) 행정동의 분기별 매출금액 합계')
plt.xticks(rotation=0)  # X축 라벨의 회전 없이 표시
plt.legend(title='분기', labels=['1', '2', '3', '4', '5'])
plt.show()
```

두 상권의 수치최대값이 달라, 혼동 올 수 있으니 주의 요망 <br> 매출 금액대는 발달상권이 골목상권보다 더 높다.
